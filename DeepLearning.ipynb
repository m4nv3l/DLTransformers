```python
from graphviz import Digraph
```

```python
dot = Digraph(comment='Single Neuron Perceptron')

# Input Nodes
dot.node('X1', 'x1')
dot.node('X2', 'x2')
dot.node('X3', 'x3')

# Weight Nodes (represented as labels on edges or separate nodes)
dot.node('W1', 'w1', shape='box')
dot.node('W2', 'w2', shape='box')
dot.node('W3', 'w3', shape='box')

# Bias Node
dot.node('B', 'b', shape='box')

# Summation Node
dot.node('Z', 'z = wT * x + b', shape='circle')

# Activation Node
dot.node('A', 'a = f(z)', shape='circle')

# Connections
dot.edge('X1', 'Z', label='* w1')
dot.edge('X2', 'Z', label='* w2')
dot.edge('X3', 'Z', label='* w3')
dot.edge('B', 'Z', label='+')
dot.edge('Z', 'A', label='f()')

# Render the graph
display(dot)
```

```python
from graphviz import Digraph

dot = Digraph(comment='One Hidden Layer Neural Network')
dot.attr(rankdir='LR') # Left to right diagram
dot.attr(nodesep='0.1', ranksep='0.4') # Reduce spacing for a more compact layout

# Styles for nodes
input_node_style = {'shape': 'circle', 'style': 'filled', 'fillcolor': 'white', 'fontcolor': 'black'}
summation_node_style = {'shape': 'circle', 'style': 'filled', 'fillcolor': 'lightblue', 'fontcolor': 'black'}
activation_node_style = {'shape': 'circle', 'style': 'filled', 'fillcolor': 'lightgreen', 'fontcolor': 'black'}
bias_node_style = {'shape': 'box', 'style': 'filled', 'fillcolor': 'yellow', 'fontcolor': 'black'}

# Input Layer
with dot.subgraph(name='cluster_input') as c:
    c.attr(label='Input Layer (X)')
    # c.attr(rank='same') # Ensure nodes within this cluster are on the same rank
    c.node('X1', 'x1', **input_node_style)
    c.node('X2', 'x2', **input_node_style)
    c.node('X3', 'x3', **input_node_style)

# Hidden Layer
with dot.subgraph(name='cluster_hidden') as c:
    c.attr(label='Hidden Layer (4 Neurons)')
    # c.attr(rank='same') # Ensure nodes within this cluster are on the same rank
    # Summation Nodes (Z_h)
    c.node('ZH1', 'z_h1', **summation_node_style)
    c.node('ZH2', 'z_h2', **summation_node_style)
    c.node('ZH3', 'z_h3', **summation_node_style)
    c.node('ZH4', 'z_h4', **summation_node_style)
    # Activation Nodes (A_h) - simplified labels
    c.node('AH1', 'a_h1', **activation_node_style)
    c.node('AH2', 'a_h2', **activation_node_style)
    c.node('AH3', 'a_h3', **activation_node_style)
    c.node('AH4', 'a_h4', **activation_node_style)
    # Bias Nodes (B_h)
    c.node('BH1', 'b_h1', **bias_node_style)
    c.node('BH2', 'b_h2', **bias_node_style)
    c.node('BH3', 'b_h3', **bias_node_style)
    c.node('BH4', 'b_h4', **bias_node_style)

# Output Layer
with dot.subgraph(name='cluster_output') as c:
    c.attr(label='Output Layer (1 Neuron)')
    # c.attr(rank='same') # Ensure nodes within this cluster are on the same rank
    # Summation Node (Z_o)
    c.node('ZO1', 'z_o1', **summation_node_style)
    # Activation Node (A_o) - simplified label
    c.node('AO1', 'a_o1', **activation_node_style)
    # Bias Node (B_o)
    c.node('BO1', 'b_o1', **bias_node_style)

# Use invisible edges to explicitly align the top nodes of each cluster on the same rank
dot.edge('X1', 'ZH1', style='invis')
dot.edge('ZH1', 'ZO1', style='invis')

# Connections from Input to Hidden Sums (with W_h weights)
# Explicit weights only for the first hidden neuron for clarity
dot.edge('X1', 'ZH1', label='w_11')
dot.edge('X2', 'ZH1', label='w_21')
dot.edge('X3', 'ZH1', label='w_31')

# Connections to other hidden neurons without explicit weight labels for brevity
dot.edge('X1', 'ZH2')
dot.edge('X2', 'ZH2')
dot.edge('X3', 'ZH2')

dot.edge('X1', 'ZH3')
dot.edge('X2', 'ZH3')
dot.edge('X3', 'ZH3')

dot.edge('X1', 'ZH4')
dot.edge('X2', 'ZH4')
dot.edge('X3', 'ZH4')

# Bias Connections to Hidden Sums
dot.edge('BH1', 'ZH1', label='+')
dot.edge('BH2', 'ZH2', label='+')
dot.edge('BH3', 'ZH3', label='+')
dot.edge('BH4', 'ZH4', label='+')

# Hidden Sums to Hidden Activations
dot.edge('ZH1', 'AH1', label='f()')
dot.edge('ZH2', 'AH2', label='f()')
dot.edge('ZH3', 'AH3', label='f()')
dot.edge('ZH4', 'AH4', label='f()')

# Connections from Hidden Activations to Output Sum (with W_o weights)
dot.edge('AH1', 'ZO1', label="w'_11")
dot.edge('AH2', 'ZO1', label="w'_21")
dot.edge('AH3', 'ZO1', label="w'_31")
dot.edge('AH4', 'ZO1', label="w'_41")

# Bias Connection to Output Sum
dot.edge('BO1', 'ZO1', label='+')

# Output Sum to Output Activation
dot.edge('ZO1', 'AO1', label='f()')

display(dot)
```

```python
from sentence_transformers import SentenceTransformer, util

# MOdel
model = SentenceTransformer('all-MiniLM-L6-v2')

# TEst: Creating a small dataset
documents = [
    "The capital of France is Paris.",
    "A numeric method for solving differential equations.",
    "Python is a popular programming language for Data Science.",
    "The Sahara is the largest hot desert in the world.",
    "Machine Learning models like BERT use the Transformer architecture."
]
```

```python
# encoding
document_embeddings = model.encode(documents)

print(f" size: {document_embeddings.shape[1]}")
```

```python
# query test
query = "How do computers understand natural language?"
query_embedding = model.encode(query) # encode query
```

```python
# similarity between query and all docuemnts:
simv = util.semantic_search(query_embedding, document_embeddings, top_k=2)
```

```python
simv
```

```python
# Results:
print(f"Query: {query}\n")
for h in simv[0]:
    idx = h['corpus_id']
    score = h['score']
    print(f"Score: {score:.4f} | Document: {documents[idx]}")
```
